```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import HashingVectorizer
from gensim.models import Word2Vec
import sklearn
```

```python
res = pd.read_pickle("tweet_prepared.pkl")
res.head()
```

```python
count_vectorizer = CountVectorizer(max_df=0.9, max_features = 1000, stop_words='english')
```

```python
BOW_count_stemmed = count_vectorizer.fit_transform(res['Preproc_text'])

# Отобразим Bag-of-Words модель как DataFrame
feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(BOW_count_stemmed.toarray(), columns = feature_names)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.9, max_features = 1000, stop_words='english')
```

```python
# Создаем the Bag-of-Words модель
BOW_tfidf_stemmed = tfidf_vectorizer.fit_transform(res['Preproc_text'])

# Отобразим Bag-of-Words модель как DataFrame
feature_names = tfidf_vectorizer.get_feature_names()
pd.DataFrame(BOW_tfidf_stemmed.toarray(), columns = feature_names)
```

```python
vectorizer = HashingVectorizer(n_features=2**4, ngram_range=(1, 1), analyzer='word', binary=False, stop_words='english')
```

```python
BOW_hash_stemmed = vectorizer.fit_transform(res['Preproc_text'])
print(BOW_hash_stemmed.shape)
print(BOW_hash_stemmed.toarray())
```
